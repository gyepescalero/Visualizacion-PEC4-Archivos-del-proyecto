```python
import pandas as pd
import numpy as np

data = pd.read_csv('Indigenous.csv')
data.head()
```

```python
data["Year"]=[data["Dat"][i][0:4] for i in range(len(data["Dat"]))]
data['Con']=[data["Con"][i].replace('(','').replace(')','') for i in range(len(data["Con"]))]
```

```python
import re
names=pd.read_excel('Country_names.xlsx')
names.head()

country=[]
code=[]
Con_Times=[]
AgtId=[]
Year=[]
Reg=[]

def find_equals(c):
    l = len(c)
    for i in range(l-1):
        if c[i] != c[i+1]:
                return False
    return True



for j in range(len(names["country"])):
    nombres=[re.search(names["country"][j], data["Con"][i]) for i in range(len(data["Con"]))]
    if find_equals(nombres)==False:
        for k in range(len(nombres)):
            if nombres[k]!=None: 
                country.append(names["country"][j]) 
                code.append(names["code"][j]) 
                Con_Times.append(data["Con"][k])                                                        
                AgtId.append(data["AgtId"][k])  
                Year.append(data["Year"][k])
                Reg.append(data["Reg"][k])
                

correct_names = {"Country": country , "Loc1ISO": code , "Con_Times": Con_Times, "AgtId": AgtId, "Year": Year,"Region": Reg}

CN=pd.DataFrame(correct_names)
CN
```
